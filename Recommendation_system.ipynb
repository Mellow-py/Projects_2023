## Import library to import data

```python
#Import library
import pandas as pd

metadata = pd.read_csv('movies_metadata.csv')
metadata.head(5)
```

```python
metadata.info()
```

```python
'''The aim of this modeling is to create a system that gives the top 250 movies on IMDB
movies chart using different metrics as given in the column above.'''
```

```python
'''The weighted rating formula is used to derive the rating and its given below: 
WR = ((v/(v+m)) * R + (m(v+m)) * C )
Where v: number of votes for a movie| m: minimum votes required to be listed on the chart|
C: mean of the vote across the whole report'''
```

```python
'''From the dataset, v="vote_count", R="vote_average", the C (average of the votes)
can be calculated from the mean of the vote_average'''
```

```python
#Calculating the value of C
C = metadata["vote_average"].mean()
print(C)
```

```python
'''The next paramater to calculate is the m(minimum votes required) using the pandas quantile
method with "the vote_count"'''

#calculating the m
m = metadata["vote_count"].quantile(.90)
print(m)
```

```python
'''Since the value for the least rating as been set to 160, a new librabry is created to 
save the movies that is higher than the value set by the quantile using the >='''

#flter out all quantified movies and save them in a new dataframe
new_movies = metadata.copy().loc[metadata['vote_count'] >= m]
new_movies.shape
```

```python
metadata.shape
```

```python
'''From the above output, there are about just 10% of the metadata dataset that are higher 
than the 90% quantile. We now calculate the weighted ratings fromA the formular'''

#A function that computes the weighted rating(WR) of each movie
def WR(x, m=m, C=C):
    v = x["vote_count"]
    R = x["vote_average"]
    #Implement the IMDB formula
    return (v/(v/m) * R + (m/(v+m)) * C)
```

```python
'''Now we define a new dataframe in the "new_movies" dataset and calculate the value with the 
function of WR()'''

#New score dataframe
new_movies["score"] = new_movies.apply(WR, axis=1)

#Sort the new_movies dataset using the "score" column in ascending order
new_movies = new_movies.sort_values('score', ascending=False)
new_movies.head(20)
```
